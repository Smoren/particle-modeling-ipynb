```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def draw_plot_lj(x, y, ax):
    plot = sns.lineplot(
        x=x,
        y=y,
        ax=ax,
    )
    plot.set_ylim(-1.5, 2.5)
    plot.axhline(y=0, color='grey')

def draw_plot_lj_truncated(x, y, ax, sigma):
    plot = sns.lineplot(
        x=x,
        y=y,
        ax=ax,
    )
    plot.set_ylim(-0.1, 0.1)
    plot.axhline(y=0, color='grey')
    plot.axvline(x=2.5*sigma, color='grey')
```

```python
def morse_potential(r: np.ndarray, eps: float, alpha: float, sigma: float):
    """
    Morse potential
    """
    return eps * (np.exp(-2*alpha * (r - sigma)) - 2*np.exp(-alpha * (r - sigma)))

def morse_force(r: np.ndarray, eps: float, alpha: float, sigma: float):
    """
    Morse force (Derivative of Morse potential)
    """
    return -2*alpha * eps * np.exp(alpha*(sigma-r)) * (np.exp(alpha*(sigma-r)) - 1)

def morse_potential_alt_form(r: np.ndarray, alpha: float, de: float, re: float):
    """
    Morse potential (alternative form)
    """
    return de * (1 - np.exp(-alpha*(r-re)))**2
```

```python
_, ax = plt.subplots(ncols=1, nrows=1, figsize=(16, 8))

x = np.linspace(0, 10, 1000)

plot = sns.lineplot(
    x=x,
    y=morse_potential(x, 1, 1, 1),
    ax=ax,
)
plot.axhline(y=0, color='grey')
plot.axvline(x=0, color='grey')

plot = sns.lineplot(
    x=x,
    y=morse_force(x, 1, 1, 1),
    ax=ax,
)
```

```python
_, axes = plt.subplots(ncols=2, nrows=1, figsize=(20, 8))

x = np.linspace(0, 10, 1000)

plot = sns.lineplot(
    x=x,
    y=morse_potential(x, 0.5, 1, 1),
    ax=axes[0],
)
sns.lineplot(
    x=x,
    y=morse_potential(x, 1, 1, 1),
    ax=axes[0],
)
sns.lineplot(
    x=x,
    y=morse_potential(x, 1.5, 1, 1),
    ax=axes[0],
)
plot.axhline(y=0, color='grey')
plot.axvline(x=0, color='grey')

plot = sns.lineplot(
    x=x,
    y=morse_force(x, 0.5, 1, 1),
    ax=axes[1],
)
sns.lineplot(
    x=x,
    y=morse_force(x, 1, 1, 1),
    ax=axes[1],
)
sns.lineplot(
    x=x,
    y=morse_force(x, 1.5, 1, 1),
    ax=axes[1],
)
plot.axhline(y=0, color='grey')
plot.axvline(x=0, color='grey')
plot.set_ylim(-5, 2);
```
